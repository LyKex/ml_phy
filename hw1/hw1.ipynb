```python
import pandas as pd
import numpy as np
```

# 1
The csv data file is downloaded from https://mlphysics.ics.uci.edu/data/susy/. The column names are determined according to readme description file.

# 2

```python
names = ["label"] + ["features_{}".format(i) for i in range(18)]
df = pd.read_csv("SUSY.csv.gz", index_col=None, header=None, names=names)
df
```

There are 18 features, and 5000000 data points. The first column is the label.

# 3

```python
df.loc[df["label"] == 0.0, "label"] = -1
```

```python
df
```

# 4

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size=0.4, random_state=42)
test, validation = train_test_split(test, test_size=0.5, random_state=42)
```

```python
print(len(train))
print(len(test))
print(len(validation))
```

```python
X_train = train.iloc[:, 1:]
y_train = train["label"]

X_test = test.iloc[:, 1:]
y_test = test["label"]

X_validation = validation.iloc[:, 1:]
y_validation = validation["label"]
```

# 5

If fitted on test and validation set, some of the distribution of the data will be known to the model. This is a data leakage and may cause overtraining and overestimation of the model performance.

```python
from sklearn.preprocessing import StandardScaler


scaler = StandardScaler()
scaler.fit(X_train)
X_train_std = scaler.transform(X_train)
X_test_std = scaler.transform(X_test)
X_validation_std = scaler.transform(X_validation)
```

```python
X_train_std.mean(axis=0)
```

```python
X_train_std.std(axis=0)
```

# 6


```python
reg_strength = 0.1

def class_accuracy(y_true, y_pred):
    return np.mean(y_true == y_pred)
```

```python
def ridge_classification(X, y, reg_strength):
    return np.linalg.inv(X.T @ X + reg_strength * np.eye(X.shape[1])) @ X.T @ y

def ridge_predict(X, w):
    return np.sign(X @ w)
```

# 7

```python
w_ridge = ridge_classification(X_train_std, y_train, reg_strength)

acc_ridge = class_accuracy(y_test, ridge_predict(X_test_std, w_ridge))
print("accuracy of ridge regression on test set: {}".format(acc_ridge))
```

# 8

`sklearn` formulation defines the labels $y_i \in \{0, 1\}$, so the conditional probability is expressed as 
$$
\begin{aligned}
p(y_i = 1| X_i, w) = \frac {\exp(y_i (X_i w))} {\exp 0+\exp(X_i w)} = \frac 1 {1+\exp(-X_i w)} \\
p(y_i = 0| X_i, w) = \frac {\exp(y_i (X_i w))} {\exp 0+\exp(X_i w)} = \frac 1 {1+\exp(X_i w)} 
\end{aligned}
$$
and max likihood is equivalent to minimize negative log likihood, which is extacly how the logistic loss function is designed
$$
l(y_i | X_i, w) = -\ln p(y_i| X_i, w) = -y_i \ln (1+\exp(-X_i w))- (1-y_i) \ln(1 + \exp(X_i w))
$$

The labels in lecture takes $y'_i \in \{-1, 1\}$, but they all use the same probaility expression (logit function) and loss function definition (maximum log likihood).

# 9

```python
y_train_sk = (y_train + 1) * 0.5
y_test_sk = (y_test + 1) * 0.5
y_validation_sk = (y_validation + 1) * 0.5
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(C = 1/reg_strength, fit_intercept=False, penalty="l2")
logreg.fit(X_train_std, y_train_sk)
acc_logreg = class_accuracy(y_test_sk, logreg.predict(X_test_std))
print("accuracy of logistic regression on test set: {}".format(acc_logreg))
```

# 10

```python
lambdas = np.logspace(-1, 4, 10)
accs_ridge_val = []
accs_ridge_train = []
accs_logreg_val = []
accs_logreg_train = []
for l in lambdas:
    w_ridge = ridge_classification(X_train_std, y_train, l)
    acc_ridge_train = class_accuracy(y_train, ridge_predict(X_train_std, w_ridge))
    acc_ridge_val = class_accuracy(y_validation, ridge_predict(X_validation_std, w_ridge))

    logreg = LogisticRegression(C = 1/l, fit_intercept=False, penalty="l2")
    logreg.fit(X_train_std, y_train_sk)
    acc_logreg_train = class_accuracy(y_train_sk, logreg.predict(X_train_std))
    acc_logreg_val = class_accuracy(y_validation_sk, logreg.predict(X_validation_std))

    print(">>> lambda: ", l)
    print("ridge classification train accuracy: ", acc_ridge_train)
    print("ridge classification validation accuracy: ", acc_ridge_val)
    print("logistic regression train accuracy: ", acc_logreg_train)
    print("logreg regression validation accuracy: ", acc_logreg_val)
    accs_ridge_val.append(acc_ridge_val)
    accs_ridge_train.append(acc_ridge_train)
    accs_logreg_val.append(acc_logreg_val)
    accs_logreg_train.append(acc_logreg_train)
```

```python
import matplotlib.pyplot as plt

plt.plot(lambdas, accs_ridge_val, label="ridge validation accuracy")
plt.plot(lambdas, accs_ridge_train, "--", label="ridge train accuracy")
plt.plot(lambdas, accs_logreg_val, label="logistic regression validation accuracy")
plt.plot(lambdas, accs_logreg_train, "--", label="logistic regression train accuracy")
plt.xscale("log")
plt.legend()
plt.show()
```

```python
id_ridge = np.argmax(accs_ridge_val)
print("best ridge lambda: ", lambdas[id_ridge])
id_logreg = np.argmax(accs_logreg_val)
print("best logistic regression lambda: ", lambdas[id_logreg])
```

Validation loss is a better indication of the model performance, since it better describes the model predictive ability on unseen data.

Thus, $\lambda =2783$ is the best value of the range for logistic regression. Although since it yields highest validation accuracy. The ridge regression shows low sensitity to the regularization strength in this case. The optimal value in the range tested is 0.1 but a wider range test may be needed.
